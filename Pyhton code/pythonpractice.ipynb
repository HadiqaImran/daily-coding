## **Problem 3 – Student Performance Report (6 marks)**
Build a grade management program that stores students' marks in multiple subjects. Display each student's average marks, and find the top 3 performers.


```python
def calc_avg():
    students = {
        ('Arij', 'Math'): 95,
        ('Arij', 'Science'): 88,
        ('Bisma', 'Math'): 78,
        ('Bisma', 'Science'): 92,
        ('Maya', 'Math'): 65,
        ('Maya', 'Science'): 40,
        ('Hadiqa', 'Math'): 88,
        ('Hadiqa', 'Science'): 99
    }

    total_math = 0
    total_science = 0
    count_math = 0
    count_science = 0
    sums = {}

    for key, marks in students.items():
        name = key[0]
        subject = key[1]

        if subject == "Math":
            total_math += marks
            count_math += 1
        elif subject == "Science":
            total_science += marks
            count_science += 1
        if name not in sums:
            sums[name] = marks
        else:
            sums[name] += marks

    avg_math = total_math / count_math
    avg_science = total_science / count_science

    print("Average Marks:")
    print("Math:", avg_math)
    print("Science:", avg_science)
    print()

    student_list = []
    for name, total in sums.items():
        student_list.append((name, total))

    for i in range(len(student_list)):
        for j in range(i + 1, len(student_list)):
            if student_list[j][1] > student_list[i][1]:
                student_list[i], student_list[j] = student_list[j], student_list[i]

    print("Top 3 Students:")
    for i in range(3):
        print(student_list[i][0], ":", student_list[i][1])

calc_avg()
```

Build a public transport tracking tool where each stop has a name, time, and distance from the start. Find the fastest segment, total route time, and total number of stops.

```python
# Build a public transport tracking tool where each stop has a name, time, and distance from the start. Find the fastest segment, total route time, and total number of stops.
def tool():
    stops=[
    {"name": "Canal Station", "time": 0, "distance": 0},
    {"name": "Kalma Chowk", "time": 15, "distance": 10},
    {"name": "Barkat Market", "time": 10, "distance": 5},
    {"name": "Gate 4", "time": 20, "distance": 13},
    ]
    total_stops = len(stops)
    total_time = stops[-1]["time"] - stops[0]["time"]
    fastest_speed = 0
    fastest_segment = None
    for i in range(1, total_stops):
        time_diff = stops[i]["time"] - stops[i - 1]["time"]
        dist_diff = stops[i]["distance"] - stops[i - 1]["distance"]
        if time_diff > 0:  
            speed = dist_diff / time_diff
            if speed > fastest_speed:
                fastest_speed = speed
                fastest_segment = (stops[i - 1]["name"], stops[i]["name"])

    print("Total number of stops:", total_stops)
    print("Total route time:", total_time, "minutes")
    print(f"Fastest segment: {fastest_segment[0]} to {fastest_segment[1]} with speed {fastest_speed}")
tool()
    
```

Develop a program to manage a sports tournament. Track teams, matches, and scores, and generate the winning team, average score, and unique participants.

```python
def tournament_manager():
   
    teams = {
        "Tigers": ["Anaya", "Abeeha", "Kashaf"],
        "Lions": ["Maha", "Eman", "Zoha"],
        "Eagles": ["Anusha", "Hadiqa", "Laiba"]
    }

    
    matches = [
        {"team1": "Tigers", "team2": "Lions", "score1": 50, "score2": 45},
        {"team1": "Tigers", "team2": "Eagles", "score1": 40, "score2": 55},
        {"team1": "Lions", "team2": "Eagles", "score1": 30, "score2": 50}
    ]

   
    total_scores = {}
    for team in teams:
        total_scores[team] = 0

    for match in matches:
        total_scores[match["team1"]] += match["score1"]
        total_scores[match["team2"]] += match["score2"]

    
    winner = None
    highest_score = 0
    for team, score in total_scores.items():
        if score > highest_score:
            highest_score = score
            winner = team

    
    total_points = sum([match["score1"] + match["score2"] for match in matches])
    total_games = len(matches) * 2
    avg_score = total_points / total_games

    participants = set()
    for player_list in teams.values():
        participants.update(player_list)

    

    print("Winning Team:", winner)
    print("Total Scores per Team:", total_scores)
    print("Average Score:", avg_score)
    print("Unique Participants:", participants)

tournament_manager()
```

## **Problem 12 – Fitness Tracker (6 marks)**
Design a program to track multiple users' fitness activities. Calculate total calories burned, average workout time, and display unique activity types.

```python
def tracker():
    users = {
        "Hadiqa": [("Running", 30, 300), ("Yoga", 45, 150), ("Cycling", 20, 200)],
        "Ali": [("Running", 40, 400), ("Swimming", 30, 250)],
        "Fahad": [("Cardio", 60, 200), ("Cycling", 25, 180), ("Running", 35, 320)]
    }

    user_calories = {}
    user_time = {}
    all_activities = set()

    for person, workouts in users.items():
        user_calories[person] = 0
        user_time[person] = 0
        for activity_name, minutes, calories_burned in workouts:
            user_calories[person] += calories_burned
            user_time[person] += minutes
            all_activities.add(activity_name)

    avg_workout_time = {}
    for person, workouts in users.items():
        avg_workout_time[person] = user_time[person] / len(workouts)


    for person, total_cal in user_calories.items():
        print(person, ":", total_cal, "cals")

    print("\nAverage Workout Time per User:")
    for person, avg_time in avg_workout_time.items():
        print(person, ":", (avg_time), "mins")

    print("\nUnique Activity Types:")
    print(all_activities)

tracker()
```

## **Problem 16 – Integrated Real-World System: Smart Business Manager (10 marks)**
Develop a 'Smart Business Manager' that combines all data-handling skills.

Your program should:
- Record business transactions with details like date, type, amount, and category
- Maintain all transactions for reporting
- Summarize totals by category
- Identify the highest expense
- Generate a formatted financial summary report
- Use modular functions such as `add_transaction()`, `total_summary()`, `highest_expense()`, etc.


```python
def add_transaction(transactions):
    date = input("Enter date: ")
    t_type = input("Enter type: ")
    amount = float(input("Enter amount: "))
    category = input("Enter category: ")
    transactions.append([date, t_type, amount, category])

def total_summary(transactions):
    totals = {}
    for t in transactions:
        category = t[3]
        amount = t[2]
        if category in totals:
            totals[category] += amount
        else:
            totals[category] = amount

    for category, total in totals.items():
        print(f"{category}: {total}")
    print()

def highest_expense(transactions):
    highest = 0
    highest_category = ""
    for t in transactions:
        if t[2] > highest:
                highest = t[2]
                highest_category = t[3]
    print(f"Highest expense: {highest_category} ({highest})\n")

def report(transactions):
    print("\nFinancial Summary Report:")
   
    for t in transactions:
        print(f"{t[0]} {t[1]}  {t[2]}  {t[3]}")
    total_summary(transactions)
    highest_expense(transactions)

def manager():
    transactions = []
    while True:
        print("1. Add Transaction")
        print("2. Show Financial Report")
        print("3. Exit")
        choice = input("Enter your choice: ")
        if choice == "1":
            add_transaction(transactions)
        elif choice == "2":
            report(transactions)
        elif choice == "3":
            break
        else:
            print("Try again.")

manager()
```
